# Dummy Variables
Handling Categorical Data in Machine Learning Models: 
Many machine learning algorithms (e.g., linear regression, logistic regression, decision trees) cannot work directly with categorical data (non-numeric). 
They require features to be represented as numbers. 
Dummy variables convert categorical features into a binary (0 or 1) numeric format so that models can process them.

```python
df=pd.get_dummies(df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.metrics import accuracy_score, classification_report, mean_squared_error

# Step 1: Create a target variable for classification
# Categorize students' performance based on Math, Reading, and Writing scores into Low, Medium, High
df['AvgScore'] = df[['MathScore', 'ReadingScore', 'WritingScore']].mean(axis=1)
df['PerformanceCategory'] = pd.cut(df['AvgScore'], bins=[0, 60, 80, 100], labels=['Low', 'Medium', 'High'])

# Step 2: Define features and target for classification
X_class = df.drop(columns=['MathScore', 'ReadingScore', 'WritingScore', 'AvgScore', 'PerformanceCategory'])
y_class = df['PerformanceCategory']

# Split the data into train and test sets for classification
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(X_class, y_class, test_size=0.2, random_state=42)

# Step 3: Train a logistic regression model for classification
clf = LogisticRegression(max_iter=1000)
clf.fit(X_train_class, y_train_class)

# Step 4: Make predictions for classification
y_pred_class = clf.predict(X_test_class)

# Step 5: Evaluate classification model
classification_report_res = classification_report(y_test_class, y_pred_class)

# Step 6: Define features and target for regression
X_reg = df.drop(columns=['MathScore', 'ReadingScore', 'WritingScore', 'AvgScore', 'PerformanceCategory'])
y_reg = df['AvgScore']

# Split the data into train and test sets for regression
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

# Step 7: Train a linear regression model for regression
reg = LinearRegression()
reg.fit(X_train_reg, y_train_reg)

# Step 8: Make predictions for regression
y_pred_reg = reg.predict(X_test_reg)

# Step 9: Evaluate regression model
mse_reg = mean_squared_error(y_test_reg, y_pred_reg)
print(y_pred_reg)
classification_report_res
```
